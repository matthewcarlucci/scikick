```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)
ax.grid()
plt.show()
```

```python
fig.savefig("output/test.png")
```
